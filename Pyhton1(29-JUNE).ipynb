```python
#Lists :
num = [1,2,3,4,5]
print(type(num))
num.append(6)
print(num)
```

```python
num = [1,2,3,4,5]
num1 = [5,8,9,10]
print(type(num))
num.append(6)
print(num)
num3 = num + num1
print(num3)
#REMOVING DUPLICATE ELEMENTS
new_list = []
for item in num3:
    if item not in new_list:
        new_list.append(item)

print(new_list)
```

```python
#CALCULATOR :
a = int(input("ENTER 1st NUMBER :"))
b = int(input("ENTER 2nd NUMBER :"))
inp = int(input("""SELECT ANY ONE OPTION :
(1)ADDITION
(2)SUBTRACTION
(3)MULTIPLICATION
(4)DIVISION
(5)REMAINDER
(6)PERCENTAGE
"""))
if (inp==1):
    print("SUM IS:",a+b)
elif(inp==2):
    print("DIFFERENCE IS:",a-b)
elif(inp==3):
    print("PRODUCT IS:",a*b)
elif(inp==4):
    print("QUOTIENT IS:",a/b)
elif(inp==5):
    print("REMAINDER IS:",a%b)
elif(inp==6):
    print("PERCENTAGE IS:",(a/b)*100)
else :
    print("INAVLID INPUT")
```
