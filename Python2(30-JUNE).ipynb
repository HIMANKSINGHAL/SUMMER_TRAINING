```python
principle = int(input("Enter Principle Amount :"))
time = int(input("Enter Time Period (in years) :"))
interest = int(input("Enter Rate of interest(per annum) :"))
simple_interest = (principle*time*interest)/100
compound_interest = (principle*(1+(interest/100))**time)-principle

print("SIMPLE INTEREST IS :",simple_interest)
print("COUMPOUND INTEREST IS:",compound_interest)
```

```python
rows = int(input("Enter the row size for the pattern: "))
for i in range(1, rows + 1): 
    for j in range(rows  - i): 
        print(" ", end=" ")
    for k in range(1, 2 * i): 
        print("*", end=" ")
    print()
```
