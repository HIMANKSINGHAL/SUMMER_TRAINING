```python
#STUDENT INFORMATION SYSTEM : CREATE STUDENT CLASS AND ADD ATTRIBUTES LIKE ROLLNO. , NAME AND MARKS(MATHS,PHYSICS,CHEMISTRY)
class Student:

    def get_Data(a):
        a.Stu_Name = input("Enter Student Name: ")
        a.Stu_Rollno = int(input("Enter Student Roll No.: "))
        a.Maths_Marks = int(input("Enter Marks in Mathematics: "))
        a.Phy_Marks = int(input("Enter Marks in Physics: "))
        a.Chem_Marks = int(input("Enter Marks in Chemistry: "))

    def display_Data(a):
        print("\n----- Student Details -----")
        print("Name :", a.Stu_Name)
        print("Roll No. :", a.Stu_Rollno)
        print("Mathematics :", a.Maths_Marks)
        print("Physics :", a.Phy_Marks)
        print("Chemistry :", a.Chem_Marks)

    def Calculate_Percentage(a):
        Total_Marks = a.Maths_Marks + a.Phy_Marks + a.Chem_Marks
        Percentage = (Total_Marks / 300) * 100
        print("Total Marks :", Total_Marks)
        print("Percentage :", Percentage, "%")
```

```python
stu1 = Student()
stu1.get_Data()
stu1.display_Data()
stu1.Calculate_Percentage()
```

```python
class BankAccount:

    def get_Data(b):
        b.Name = input("Enter Account Holder Name: ")
        b.AccountNumber = int(input("Enter Account Number: "))
        b.Balance = 25000

    def display_Data(b):
        print("\nAccount Information")
        print("Account Holder Name :", b.Name)
        print("Account Number :", b.AccountNumber)

    def deposit(b):
        D_Amt = int(input("Enter Amount To Deposit: "))
        b.Balance += D_Amt
        print("Amount Deposited Successfully.")

    def withdraw(b):
        W_Amt = int(input("Enter Amount To Withdraw: "))
        if W_Amt <= b.Balance:
            b.Balance -= W_Amt
            print("Amount Withdrawn Successfully.")
        else:
            print("Insufficient Balance!")

    def display_Balance(b):
        print("Current Balance :", b.Balance)

    def Operations(b):
        while True:
            choice = int(input("""
Enter Your Choice:
1) Deposit
2) Withdraw
3) Display Balance
4) Exit
Choice: """))

            if choice == 1:
                b.deposit()
            elif choice == 2:
                b.withdraw()
            elif choice == 3:
                b.display_Balance()
            elif choice == 4:
                print("Thank You!")
                break
            else:
                print("Invalid Choice!")
```

```python
b1 = BankAccount()
b1.get_Data()
b1.display_Data()
b1.Operations()
```
